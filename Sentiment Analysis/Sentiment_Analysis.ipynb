```python
import pandas as pd
import urllib.request as request
import seaborn as sns
```

```python
lex = request.urlopen('https://raw.githubusercontent.com/theleadio/datascience_demo/master/OpinionFinder_Lexicon.tff').read().decode('UTF-8').splitlines()
```

```python
lexicon = {}

for line in lex:

  #split elements
  tokens = line.split(' ')

  #get the word
  word = tokens[2].replace('word1=','')

  #get the priority
  prior = tokens[-1].replace('priorpolarity=', '')
  if  prior == 'negative': polarity = -1
  elif prior == 'positive': polarity = 1
  else: polarity = 0

  #get the strength
  strength = tokens[0].replace('type=', '')
  if  strength == 'strongsubj': s = 3
  elif strength == 'weaksubj': s = 1
  else: s = 0

  lexicon[word] = polarity * s
```

```python
lexicon['shiok'] = 5
```

```python
def sentiment(s):
  tokens = s.lower().split()

  score=0
  for token in tokens:
    if token in lexicon:
      score = score + lexicon[token]

  return score
```

```python
sentiment('I am not shiok and happy')
```

```python
df = pd.read_csv('Womens Clothing E-Commerce Reviews.csv')[['Review Text', 'Rating']].dropna()
df.head()
```

```python
sentiment_scores = [sentiment(review) for review in df['Review Text']]
sentiment_scores
```

```python
df['Sentiment Score'] = sentiment_scores
df
```
